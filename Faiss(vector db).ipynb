```python
import pandas as pd

pd.set_option('display.max_colwidth', 100)
```

```python
df = pd.read_csv('sample_text.csv')
df.shape
```

```python
df
```

```python
## Converting the sentences into vectors

from sentence_transformers import SentenceTransformer

encoder = SentenceTransformer("all-mpnet-base-v2") ## using all mpnet entity
vectors = encoder.encode(df.text)
vectors.shape
```

```python
vectors
```

```python
dim = vectors.shape[1]
dim
```

```python
import faiss

##Creating an index that will allow to do faster search

index = faiss.IndexFlatL2(dim) ##IndexFlatL2 is the index that uses eucledian distance or L2 distance
index ## Just creates an empty index
```

```python
## Adding some vectors in that empty faiss index

index.add(vectors)
```

```python
## Giving an input search query

search_query="I want to buy a neck tie"

vec = encoder.encode(search_query)
vec.shape ##Gives size 768 but expects a 2d array
```

```python
vec
```

```python
import numpy as np
svec = np.array(vec).reshape(1,-1)
svec.shape
```

```python
svec
```

```python
distances, I = index.search(svec, k=2) # 2 similar vectors, 
## it returns a tuple, one is the distances, 
## the second one is the index in our original df 
```

```python
I
```

```python
df.loc[(I[0])]
```
